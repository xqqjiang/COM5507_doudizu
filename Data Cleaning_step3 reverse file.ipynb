# Reverse the content order of all.csv(regions data in Shenzhen)  #

```python
#! /usr/bin/python

# To execute, type the following in command line:
# python ./csv-reverse-rows.py

import os
import csv


def readRows(fileName):
    # Open up the CSV file
    inCSV = open(fileName)
    reader = csv.reader(inCSV, delimiter=',')

    # Read the rows
    rows = []
    for row in reader:
        rows.append(row)

    # Close the file
    inCSV.close()

    return rows


def reverseRows(rows):
    # Separate titles and contents
    titles = [rows[0]]
    contents = rows[1:]

    # Reverse the contents
    contents.reverse()

    # Fix contents row indices
    for index in range(len(contents)):
        contents[index][0] = str(index)

    # Concatenate titles and contents
    reversedRows = titles + contents

    return reversedRows


def writeRows(fileName, rows):
    # Open up the file to write
    outCSV = open(fileName, mode='w')
    writer = csv.writer(outCSV, delimiter=',')

    # Write the rows
    for row in rows:
        writer.writerow(row)

    # Close the files
    outCSV.close()


def generateFileName(fileName, fileExtension, suffix):
    # Omit file extension and add suffix
    customName = fileName[:-len(fileExtension)]
    customName += suffix

    newFileName = customName + fileExtension

    return newFileName


# <<<< Main Procedure >>>>

# Set up the information of files to process
fileNames = [
    "./all.csv",
   
]
fileExtension = ".csv"
resultSuffix = "-reversed"

# Complete the task
print("==== _ csv-reverse-rows _ ====")
print("current directory: {}".format(os.getcwd()))

for fileName in fileNames:
    print("processing file: {}".format(fileName))

    rows = readRows(fileName)
    reversedRows = reverseRows(rows)

    newFileName = generateFileName(fileName, fileExtension, resultSuffix)
    writeRows(newFileName, reversedRows)

    print("processing complete; written to file: {}".format(newFileName))

print("==== ^ csv-reverse-rows ^ ====")
```
